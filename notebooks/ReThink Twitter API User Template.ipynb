```python
# importing functions from rethink_twitter_functions.py
from rethink_twitter_functions import *

# defining some search strings for the API queries
query1 = 'omicron'
query1_label = "Omicron Variant"
query1_start = "12/1/2021"
query1_end = "12/9/2021"
```

## 7-day search

```python
query1_7 = search_7(query1, max_results=500, write_csv=False)
query1_7.head()
```

```python
# displaying the top influencers for the 7-day search
top_influencers_7 = top_influencers(query1_7)
top_influencers_7.head()
```

## 30-day search

```python
query1_30 = search_30(query1, max_results=100, write_csv=False, label="30day")
query1_30.head()
```

```python
# displaying the top influencers for the 30-day search
top_influencers_30 = top_influencers(query1_30)
top_influencers_30.head()
```

## Full archive search

```python
query1_full = search_full(query1, start_date=query1_start, end_date=query1_end, max_results=100, write_csv=False, label="full")
query1_full.head()
```

```python
# displaying the top influencers for the full archive search
top_influencers_full = top_influencers(query1_full)
top_influencers_full.head()
```

## Wordclouds

```python
wordcloud, hashtags = word_cloud(query1_full, query=query1, save_imgs=False)
```
